# Importing dataset 

```python
df = pd.read_csv('D:/Internshala/Python/Project/employee_data.csv', header =0)
cities =pd.read_csv('D:/Internshala/Python/Project/cities.csv',header =0)
colleges =pd.read_csv('D:/Internshala/Python/Project/Colleges.csv',header =0)
```

```python
df.head()
```

##  Converting College and cities to list

```python
Tier1 = colleges["Tier 1"].tolist()
Tier2 = colleges["Tier 2"].tolist()
Tier3 = colleges["Tier 3"].tolist()
metro = cities["Metrio City"].tolist()
non_metro = cities["non-metro cities"].tolist()
```


## Converting College and cities column from categorical to numerical data type

```python
for item in df.College:
    if item in Tier1:
        df["College"].replace(item,1,inplace =True)
    elif item in Tier2:
        df["College"].replace(item,2,inplace = True)
    elif item in Tier3:
        df["College"].replace(item,3,inplace =True)
```

```python
for item in df.City:
    if item in metro:
        df["City"].replace(item,1, inplace =True)
    elif item in non_metro:
        df["City"].replace(item,0, inplace = True)
```

## creating dummies for categorical column

```python
df = pd.get_dummies(df, drop_first =True)
```

## Ridge Regression

#### Standardizing data

```python
from sklearn import preprocessing
```

#### Scaler object to store scaling info

```python
scaler = preprocessing.StandardScaler().fit(X_train)
```

```python
X_train_s = scaler.transform(X_train)
X_test_s =scaler.transform(X_test)
```

### Ridge

```python
from sklearn.linear_model import Ridge
```

```python
lm_r = Ridge(alpha = 0.8)
```

```python
lm_r.fit(X_train_s,y_train)
```

#### Evaluating R2 score

```python
r2_score(y_test,lm_r.predict(X_test_s))
```

### Validation curve to find the value of alpha for which R2 is maximum

```python
from sklearn.model_selection import validation_curve
```

```python
params = np.logspace(-2,8,100)
```

```python
train_scores, test_scores = validation_curve(Ridge(),X_train_s,y_train,param_name='alpha',param_range=params,scoring='r2')
```

```python
print(train_scores, test_scores)
```

```python
train_mean = np.mean(train_scores, axis = 1)
test_mean = np.mean(test_scores, axis =1)
```

```python
max(train_mean)
```

```python
max(test_mean)
```

```python
sns.jointplot(x=np.log(params), y=test_mean)
```

#### Finding the location where our model have maximum efficiency

```python
np.where(test_mean ==max(test_mean))
```

```python
params[34]
```

#### Fitting model and finding R2 value

```python
lm_r_best = Ridge(alpha = params[34])
```

```python
lm_r_best.fit(X_train_s,y_train)
```

```python
r2_score(y_test,lm_r_best.predict(X_test_s))
```
